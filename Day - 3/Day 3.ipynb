```python
# Control flow statements in Python means the order in which individual statements, instructions or function calls are executed or evaluated. 
# The control flow statements include conditional statements (if, elif, else), loops (for, while), and control flow modifiers (break, continue, pass).
```

```python
# 1. Conditional Statements
# -----------------------------------

# a. If- else Statement:
age = 20

if age >= 18:
    print("You are eligible to vote.")
else:
    print("You are not eligible to vote.")
```

```python
# b. Elif Statement:
score = 85

if score >= 90:
    print("Grade = 'A'")
    
elif score >= 80:
    
    print("Grade = 'B'")    
    
elif score >= 70:
    print("Grade = 'C'")
    
else:
    print("Grade = 'F'")
```

```python
# 2. Loops
# -----------------------------------

# a. For Loop: means to iterate over a sequence (like a list, tuple, dictionary, set, or string).

for i in range(5):
    print(i)
```

```python

# b. While Loop: repeatedly executes a block of code as long as a given condition is true.


count = 0
while count < 5:
    print(count)
    count += 1
```

```python
    
# 3. Control Flow Modifiers
# -----------------------------------

# a. Break Statement: is used to exit a loop prematurely when a certain condition is met.
for i in range(10):
    if i == 5:
        break
    print(i)   
```

```python
# b. Continue Statement: is used to skip the current iteration of a loop and move to the next iteration.

for i in range(10):
    if i % 2 == 0:
        continue
    print(i)
```

```python
# c. Pass Statement: is a null operation; it is used when a statement is syntactically required but you do not want any command or code to execute.
for i in range(5):
    pass  # Placeholder for future code
```

```python
# BMI Calculator
# --------------

print("Welcome to the BMI Calculator")
height=float(input("Enter your height in m :"))
weight=float(input("Enter your weight in kg :"))

BMI=round(weight/height**2)


if BMI<18.5:
    print(f"your bmi is {BMI},you have underweight")
    
elif 18.5<=BMI<25:
    print(f"your bmi is {BMI},you have a normal weight")

elif 25<BMI<30:
    print(f"your bmi is {BMI},you have a overweight")

elif 30<BMI<35:
    print(f"your bmi is {BMI},you are Obese")

else:
    print(f"your bmi is {BMI},you are clinically obsese")
    
    
```

```python
# Leap year checker

print("Welcome to leap year checker.")
year = int(input("Enter the year: "))

if year % 4 == 0:
    # Divisible by 4
    if year % 100 == 0:
        # If divisible by 100, must also be divisible by 400 to be a leap year
        if year % 400 == 0:
            print(f"{year} is a leap year.")
        else:
            print(f"{year} is an ordinary year / not a leap year.")
    else:
        print(f"{year} is a leap year.")
else:
    print(f"{year} is an ordinary year / not a leap year.")
    
    
    
    
```

```python
# Roller Coaster 
# -----------

 
print("Welcome to the Roller Coaster")
height = int(input("Enter the height in cm: "))
Bill = 0

if height >= 120:
    print("You can ride the roller coaster")
    
    age = int(input("Enter your age: "))
    
    if age > 18:
        Bill += 12
        print(f"Ticket price is $12, Total Bill = ${Bill}")
        
    elif age >= 12:
        Bill += 5
        print(f"Ticket price is $5, Total Bill = ${Bill}")
        
    else:
        Bill += 7
        print(f"Ticket price is $7, Total Bill = ${Bill}")
        
    option = input("Do you want photos? Choose 'Y' or 'N': ")
    
    if option.upper() == 'Y':
        Bill += 3
        print(f"Photo taken, $3 added. Total Bill = ${Bill}")
    else:
        print(f"No photo taken. Total Bill = ${Bill}")
        
else:
    print("You cannot ride the roller coaster")
```

```python
print("Welcome to Pizza Deliveries")

size = input("What size pizza do you want? S, M, or L: ")
add_pepperoni = input("Do you want pepperoni? Y or N: ")
extra_cheese = input("Do you want extra cheese? Y or N: ")

Bill = 0

if size.upper() == 'S':
    Bill += 15
    print(f"Small pizza costs $15, Total Bill = ${Bill}")

elif size.upper() == 'M':
    Bill += 20
    print(f"Medium pizza costs $20, Total Bill = ${Bill}")

elif size.upper() == 'L':
    Bill += 25
    print(f"Large pizza costs $25, Total Bill = ${Bill}")

# Pepperoni charges
if add_pepperoni.upper() == 'Y':
    if size.upper() == 'S':
        Bill += 2   # Small pizza pepperoni costs $2
    else:
        Bill += 3   # Medium/Large pepperoni costs $3
    print(f"Pepperoni added. Total Bill = ${Bill}")

# Extra cheese charges
if extra_cheese.upper() == 'Y':
    Bill += 1
    print(f"Extra cheese added. Total Bill = ${Bill}")

print(f"Final cost of pizza = ${Bill}")
```
