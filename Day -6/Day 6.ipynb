```python
# While loop:

# A while loop in Python is used to repeatedly execute a block of code as long as a given condition is true.
i=0
while i<10:
    print(i)
    i+=1

```

```python
# For loop:

# A for loop in Python is used to iterate over a sequence (like a list, tuple, string, or range) and execute a block of code for each element in that sequence.

for i in range(1,11): # 11 is not included.
  print(i)
```
