```python
# String Manipulation

print("Hello World!\nHello World!")   # \n is used for new line.
print("Hello " + "World!")            # + is used to concatenate two strings.
print("Hello " * 3)                   # * is used to repeat the string multiple times.
print("Hello"[0])                     # Indexing in string, 0 is the first character.
print("Hello"[1])                     # 1 is the second character.
print("Hello"+ " " + "World")      # Concatenate with space in between.
```

```python
# Varaibles
# A variable is a container for a value, which can be changed during program execution.

name=input("Enter your name:\n")
print(name)

# Length Function
length = len(name)   # len() function returns the length of the string.
print("Length of your name is:", length)
```

```python
# Variable nameing Conventions
# 1. Variable names can contain letters, numbers, and underscores.
# 2. Variable names cannot start with a number.
# 3. Variable names are case-sensitive.
# 4. Variable names should not contain spaces.

# Example of variable naming

user_name = input("Enter your name:\n")
userAge = input("Enter your age:\n")
user1_city = input("Enter your city:\n")
print("Name:", user_name)
print("Age:", userAge)
print("City:", user1_city)
# --------------------------------------------
```
