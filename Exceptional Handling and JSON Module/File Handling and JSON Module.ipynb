```python

# File Input (I): Reading data from a file into your program.

# File Output (O): Writing data from your program into a file.

# It’s how programs interact with external storage (like text files, CSVs, logs, etc.).
```

```python
# 7.open disk file for update(r + w)

f = open("sample.txt", "r+")

f.write("Hello! My name is Ankit Mahato.")

f.close()
```

```python
# 7.open disk file for update(r + w)

f = open("sample.txt", "a+")

f.write("Hello! My name is Ankit Mahato.")

f.close()
```

```python
# Word Search
line=1
word="Mahato"
with open("sample.txt","r") as f:
    while True:
        content=f.readline()
        if (word in content):
            print(f"{word} found at line:{line}!")
            break

        line +=1
        print(content)


 
```

```python
# Exception handling is a mechanism in programming to deal with unexpected errors or unusual conditions during program execution.

# Purpose: Instead of crashing, the program can catch the error, handle it gracefully, and continue running or exit cleanly.

# Try block → Code that might throw an error.

# Except/Catch block → Code that runs if an error occurs.

# Finally block → Code that always runs (cleanup, closing files, etc.).
```

```python
# Exception Handling: it helps to handle the errors in the code.

try:
    x=int(input("Enter a number: "))
    # Code that might raise an exception
    result = 10 / x

except ZeroDivisionError:
    # Handle the exception
    print("Cannot divide by zero!")

except ValueError:
    # Handle the exception
    print("Please enter a valid number!")

else:
    print("Result is:", result)

finally:
    # Code that always runs
    print("Execution completed.")


```

```python
# List Comprehension

'''
[output expression for item in iterable if condition]   '''

squres=[]

for i in range(1,11):
    squres.append(i**2)
    
print(squres)
```

```python

'''
output for iterable if Condition'''

# output =x**2

# iterable=range(1,11)


sq=[x**2 for x in range(1,11)]

print(sq)
```

```python

'''
output for iterable if Condition'''

# output =x**2

# iterable=range(1,11)

# condition=x%2!=0


sq=[x**2 for x in range(1,11) if x%2!=0]

print(sq)
```

JSON Module

```python
# JSON stands for JavaScript Object Notation.

# Purpose: It’s a text-based format used to represent structured data.

# Usage: Commonly used for transferring data between a server and a client in web applications.

# it is a lightweight format that is easy to read and write. 

# it is similar to Python dictionaries.
```

```python
# JSON   Python
# object  dict
# array   list
# string  str
# number  int/float
# boolean true/false
# null    None
```

```python
# JSON Module

# 1.json.loads: Convert JSON string to Python object
# 2.json.dumps: Convert Python object to JSON string

# 3.json.load: Read JSON file
# 4.json.dump: Write JSON file
```

```python
import json

# string format JSON
json_data = '{"name": "Ankit", "age": 20}'
print(type(json_data))

# convert string format JSON to Python object
data = json.loads(json_data)
print(data)
print(type(data))
```

```python
data = {
    "name": "Ankit",
    "age": 20,
    "city": "Noida"
}

# Convert Python object to JSON string
json_string = json.dumps(data)
print(json_string)

```
