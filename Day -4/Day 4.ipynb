```python
# Module:

# In Python, a module is essentially a file containing Python code (definitions, functions, classes, and variables).
# It's a way to group related code together and make it easier to reuse.Think of it like a building block ‚Äì you can combine several blocks to create a 
# larger, more complex structure.

# Creating a Module:

# To create a module, simply create a new Python file with a .py extension. For example, let's create a module named my_module.py:
# write pi value in my_module.py
# pi = 3.14159

# Using a Module:

# To use a module in another Python file, you can use the import statement. For example, let's create another file named main.py that uses my_module:

import my_module
radius = 5
area = my_module.pi * (radius ** 2)
print(f"The value of pi is {my_module.pi}")
print(f"The area of a circle with radius {radius} is {area}")
```

```python
# Python Lists:

# A list in Python is a collection of items that are ordered and changeable. Lists are written with square brackets [] and can contain items of different data types,
# such as integers, strings, and even other lists.


# Creating a List:
my_list = [1, 2, 3, "apple", "banana", True, 3.14]
print(my_list)

# Accessing List Items:
first_item = my_list[0]  # Accessing the first item
print(f"First item: {first_item}")

last_item = my_list[-1]  # Accessing the last item
print(f"Last item: {last_item}")

# Modifying List Items:
my_list[1] = "orange"  # Changing the second item
print(f"Modified list: {my_list}")

# Adding Items to a List:
my_list.append("grape")  # Adding an item to the end of the list
print(f"List after appending: {my_list}")

my_list.insert(2, "kiwi")  # Inserting an item at index 2
print(f"List after inserting: {my_list}")

# Removing Items from a List:
my_list.remove("apple")  # Removing an item by value
print(f"List after removing 'apple': {my_list}")

popped_item = my_list.pop()  # Removing the last item
print(f"Popped item: {popped_item}")
print(f"List after popping: {my_list}")

# List Length:
list_length = len(my_list)
print(f"Length of the list: {list_length}")

# List Slicing:
sliced_list = my_list[1:4]  # Slicing the list from index 1 to 3
print(f"Sliced list (index 1 to 3): {sliced_list}")

# Looping Through a List:

for item in my_list:
    print(f"List item: {item}")
    


```

```python
# Example: Finding the largest number in a list
numbers = [3, 5, 2, 8, 1, 4]
largest_number = numbers[0]
for number in numbers:
    if number > largest_number:
        largest_number = number
print(f"The largest number is: {largest_number}")
```

```python
# Rock Paper Scissors Game:

import random

Rock = '''
    _______
---'   ____)
      (_____)
      (_____)
      (____)
---.__(___)

'''

Paper = '''
    _______
---'   ____)____
          ______)
          _______)
         _______)
---.__________)

'''

Scissors = '''
    _______
---'   ____)____
          ______)
       __________)
      (____)
---.__(___)

'''

print("Welcome to Rock, Paper, Scissors!")

user_choice = int(input("Choose 0 for Rock, 1 for Paper and 2 for Scissors: "))
computer_choice = random.randint(0, 2)

# Show choices
choices = [Rock, Paper, Scissors]
print("You chose:")
print(choices[user_choice])
print("Computer chose:")
print(choices[computer_choice])

# Game logic
if user_choice == computer_choice:
    print("Game is tie!")

elif (user_choice == 0 and computer_choice == 2) or \
     (user_choice == 1 and computer_choice == 0) or \
     (user_choice == 2 and computer_choice == 1):
    print("You win the game!")

else:
    print("You lose the game!")
```
