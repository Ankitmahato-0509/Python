```python
# DAY2

# Datatypes: it defines what type of value a variable can hold.

# Datatypes in Python
# ------------------------------
# 1. Numbers
# 2. Strings
# 3. Lists
# 4. Tuples
# 5. Dictionaries
# 6. Sets
# ------------------------------

# 1. Numbers: it includes Integer, Float, Complex Number
a = 10          # Integer
b = 3.14        # Float
c = 2 + 3j     # Complex Number
print("Numbers:")
print("Integer:", a)
print("Float:", b)

# ------------------------------

# 2. Strings: it includes sequence of characters
s = "Hello, World!"
print("\nStrings:")
print("String:", s)
print("Length of string:", len(s))
print("Uppercase:", s.upper())
print("Lowercase:", s.lower())

# ------------------------------

# 3. Lists: it includes ordered, mutable collection of elements
my_list = [1, 2, 3, 4, 5]
print("\nLists:")
print("List:", my_list)
my_list.append(6)
print("Appended List:", my_list)
print("First Element:", my_list[0])
print("Sliced List (2-4):", my_list[1:4])
my_list.remove(3)
print("List after removal:", my_list)

# ------------------------------

# 4. Tuples: it includes ordered, immutable collection of elements
my_tuple = (1, 2, 3, 4, 5)
print("\nTuples:")
print("Tuple:", my_tuple)
print("First Element:", my_tuple[0])
print("Sliced Tuple (2-4):", my_tuple[1:4])
# Tuples are immutable, so we cannot append or remove elements
print("Length of tuple:", len(my_tuple))

# ------------------------------

# 5. Dictionaries: it includes key-value pairs
my_dict = {'name': 'Alice', 'age': 25, 'city': 'New York'}
print("\nDictionaries:")
print("Dictionary:", my_dict)
print("Name:", my_dict['name'])
my_dict['age'] = 26
print("Updated Age:", my_dict['age'])
my_dict['country'] = 'USA'
print("Added Country:", my_dict)    

# ------------------------------

# 6. Sets: it includes unordered collection of unique elements
my_set = {1, 2, 3, 4, 5}
print("\nSets:")
print("Set:", my_set)
my_set.add(6)
print("Set after adding 6:", my_set)
my_set.remove(3)
print("Set after removing 3:", my_set)
print("Is 4 in set?", 4 in my_set)

```

```python
# Example of string indexing and slicing

print("Hello"[0])
print("Hello"[1:4])
print("Hello"[-1])
print("Hello"[-4:-1])

# String Concatenation

text = "123" + "456" # it joins two strings together
print(text)
print(type(text))

# String Repetition: it repeats the string multiple times
text = "abc" * 3
print(text)


    
```

```python
# Integer = whole numbers, both positive and negative, including zero

print(123 + 456)

# Large Numbers

print(123_456_789) # underscores for better readability than using commas in numbers system like 98,234,321.


# Float = decimal numbers, representing real numbers with fractional parts
print(3.14 + 2.71)

# Floating point numbers can also be represented in scientific notation
print(314.23)
print(3.1423)

# Boolean = True or False values
is_raining = True
is_sunny = False
print(is_raining)
print(is_sunny)

 
```

```python
# Mathematical Operations

print("My Age:", str(19))
print(123+456)
print(10 - 5)
print(4 * 3)
print(6 / 4)  # Division always results in a float
print(6 // 4) # Floor Division results in an integer
print(7 % 3)   # Modulus gives the remainder
print(2 ** 3)  # Exponentiation


# PEMDAS/BODMAS Rules
print(3 + 4 * 2)        # Multiplication before Addition
print((3 + 4) * 2)      # Parentheses change the order  
print(7 + 3 * (10 / (12 / (3 + 1) - 1))) # Complex expression

# PEMDASLR

print(5 - 2 + 3)        # Left to Right for same precedence
print(5 / 2 * 4)        # Left to Right for same precedence
print(2 ** 3 ** 2)      # Right to Left for Exponentiation
print((2 ** 3) ** 2)    # Using parentheses to change order
```

```python
# Project 2 
# ------------------------------


# Tip Calculator

print("Welcome to the tip calculator!")
total_bill = float(input("What is the total bill? $"))
tip_percentage = int(input("What percentage tip would you like to give? 10, 12, or 15? "))

if tip_percentage == 12:
    tip_amount = total_bill*0.12
elif tip_percentage == 15:
    tip_amount = total_bill*0.15
else:
    tip_amount = total_bill*0.10
number_of_people = int(input("How many people to split the bill? "))
bill_per_person = total_bill / number_of_people
final_amount = round(bill_per_person + (tip_amount / number_of_people), 2)
print(f"Each person should pay: ${final_amount}")
```
